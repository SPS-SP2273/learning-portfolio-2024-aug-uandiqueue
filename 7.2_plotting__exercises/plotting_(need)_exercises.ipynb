```python
from matplotlib import pyplot as plt

# Some data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr=x_err, yerr=y_err,
             color='brown', linestyle='dashed',
             marker='o', markerfacecolor='yellow',
             ecolor='black',capsize=5,
             label='Y values with errors')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='lower right')
plt.show()
```

```python
import numpy as np
x_list = np.linspace(0, 50, 1000)
y_list = np.exp(-x_list/10)*np.sin(x_list)

plt.plot(x_list, y_list)
plt.xlabel('x', size = 15, c = 'darkorange')
plt.ylabel('y', size = 15, c = 'darkorange', rotation = 'horizontal')
plt.grid(alpha=.25)
plt.title('$y=e^{-x/10}sin(x)$', size = 20)
plt.ylim(-1, 1)
plt.show()
```
