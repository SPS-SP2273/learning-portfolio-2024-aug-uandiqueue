```python
print("[", end = "")
for number in range(10):
    if (number % 2 == 0):
        continue
    if (number == 9):
        print(f"{number**2}", end = "")
    else:
        print(f"{number**2},", end = " ")
print("]")
```

```python
%%timeit
print("[", end = "")
for number in range(10):
    if (number % 2 == 0):
        continue
    if (number == 9):
        print(f"{number**2}", end = "")
    else:
        print(f"{number**2},", end = " ")
print("]")
```

```python
import numpy as np
co2_data = [
    (2000, 369.55), (2001, 371.14), (2002, 373.28), 
    (2003, 375.80), (2004, 377.52), (2005, 379.80), 
    (2006, 381.90), (2007, 383.79), (2008, 385.60), 
    (2009, 387.43), (2010, 389.90), (2011, 391.65), 
    (2012, 393.85), (2013, 396.52), (2014, 398.65),
    (2015, 400.83), (2016, 404.24), (2017, 406.55), 
    (2018, 408.52), (2019, 411.44), (2020, 414.24)
]

np_co2 = np.array(co2_data)
co2_level = np_co2[:, 1]
for year in range(len(co2_level)):
    if (year == len(co2_level) - 1):  #required so that in the next step, the last year's [year+1] will not exceed the index limit of the list
        break
    if (co2_level[year + 1] - co2_level[year] >= 3):  #finding the two years with changes more than 3ppm
        print(f"Between {int(np_co2[year][0])} and {int(np_co2[year + 1][0])}")
        print(f"Change in CO2 concentration = {co2_level[year + 1] - co2_level[year]} ppm")
```
